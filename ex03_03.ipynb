```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
samples = np.random.normal(size=1000)
mean = np.mean(samples)
median = np.median(samples)
sd = np.std(samples)
q1 = np.percentile(samples, 25)
q2 = np.percentile(samples, 75)
```

```python
plt.hist(samples, bins=20)
plt.axvline(x=mean, c='red', label='Mean')
plt.axvline(x=median, c='black', label='Median')
plt.axvline(x=q1, c='blue', label='Interquartile')
plt.axvline(x=q2, c='blue')
plt.axvline(x=-sd, c='orange', label="Standard Dev")
plt.axvline(x=sd, c='orange')
plt.legend()
plt.show()
```

```python
samples = np.random.beta(2, 5, size=1000)
mean = np.mean(samples)
median = np.median(samples)
sd = np.std(samples)
print(sd)
q1 = np.percentile(samples, 25)
q2 = np.percentile(samples, 75)
```

```python
plt.hist(samples, bins=20)
plt.axvline(x=mean, c='red', label='Mean')
plt.axvline(x=median, c='black', label='Median')
plt.axvline(x=q1, c='blue', label='Interquartile')
plt.axvline(x=q2, c='blue')
plt.axvline(x=mean - sd, c='orange', label="Standard Dev")
plt.axvline(x=mean + sd, c='orange')
plt.legend()
plt.show()
```
