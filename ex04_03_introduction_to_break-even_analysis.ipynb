```python
import numpy as np
import matplotlib.pyplot as plt
from sympy.solvers import solve
from sympy import Symbol
```

```python
x = Symbol('x')

sols = solve(8.99 * x - 6.56 * x - 1312.13, x)
sols
```

```python
xs = np.linspace(0, 1000, 1000)
plt.plot(xs, 2.43 * xs - 1312.13)
plt.axhline(0, c='k')
plt.show()
```
