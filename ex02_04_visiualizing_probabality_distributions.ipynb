```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
samples = np.random.normal(0, 1, 3000)
x = np.linspace(samples.min(), samples.max(), 3000)
y = stats.norm.pdf(x)
```

```python
plt.hist(samples, alpha=0.2, bins=25, density=True)
plt.plot(x, y)
plt.show()
```

```python
samples = np.random.beta(2, 5, size=3000)
x = np.linspace(samples.min(), samples.max(), 3000)
y = stats.beta.pdf(x, 2, 5)
```

```python
plt.hist(samples, alpha=0.2, bins=25, density=True)
plt.plot(x, y)
plt.show()
```

```python
samples = np.random.gamma(1, size=3000)
x = np.linspace(samples.min(), samples.max(), 3000)
y = stats.gamma.pdf(x, 1)
```

```python
plt.hist(samples, alpha=0.2, bins=25, density=True)
plt.plot(x, y)
plt.show()
```
