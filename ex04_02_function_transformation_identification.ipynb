```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 1000)

plt.plot(x, x**3 - x, c='blue')
plt.plot(x, (x+2)**3 - (x+2), c='red')
plt.ylim(-5, 5)
plt.show()
```

```python
plt.plot(x, np.sin(x), c='blue') #original function
plt.plot(x, np.sin(x) * 2, c='red')
plt.ylim(-5, 5)
plt.show()
```

```python
plt.plot(x, np.sin(x), c='blue')
plt.plot(x, np.sin(x * 2), c='red')
plt.ylim(-5, 5)
plt.show()
```
